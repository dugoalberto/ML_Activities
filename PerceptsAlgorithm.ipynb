```python
import numpy as np
import matplotlib.pyplot as plt

#data points created using gpt
def create_linear_separable_data(n_samples=100):
    np.random.seed(0)  # For reproducibility
    X_class1 = np.random.randn(n_samples//2, 2) + np.array([2, 2])  # Centered around (2,2)
    X_class2 = np.random.randn(n_samples//2, 2) + np.array([-2, -2])  # Centered around (-2,-2)

    X = np.vstack((X_class1, X_class2))  # Combine class 1 and class 2 points
    y = np.hstack((np.ones(n_samples//2), -1 * np.ones(n_samples//2)))  # Labels: 1 for class1, -1 for class2

    return X, y

# Generate the data
X, y = create_linear_separable_data()

# Visualize the dataset
plt.scatter(X[y==1][:, 0], X[y==1][:, 1], color='blue', label='Class 1')
plt.scatter(X[y==-1][:, 0], X[y==-1][:, 1], color='red', label='Class -1')
plt.title('Linearly Separable Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
#percepton learning algorithm
def l_algo(X, miu):
  n = X.shape[1]
  w = np.random.randint(1, 100, (n))
  i = np.random.randint(0, n-1)
  x = X[i]
  t = X[i, -1]

  while(are_all_correct(X,w) == False):
    i = np.random.randint(1, n)
    x = X[i]
    t = X[i, -1];
    o = np.sign(np.dot(w, x))
    w = w + miu*(t-o)*x
  return w


def are_all_correct(X,w):
  for i in range(X.shape[0]):
    x = X[i]
    t = X[i, -1];
    return False if t != np.sign(np.dot(x.T, w)) else None

  return True

```

```python
def plot_decision_boundary(X, y, w):
    # Plot the data points
    plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='blue', label='Class 1')
    plt.scatter(X[y == -1][:, 0], X[y == -1][:, 1], color='red', label='Class -1')

    # Plot the decision boundary (w1*x1 + w2*x2 + b = 0)
    x1 = np.linspace(-4, 4, 100)  # x1 range for plotting
    x2 = -(w[0] / w[1]) * x1  # Solving for x2 (decision boundary)

    plt.plot(x1, x2, color='green', label='Decision Boundary')
    plt.title('Linearly Separable Data with Decision Boundary')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend()
    plt.show()
```

```python
X_with_labels = np.hstack((X, y.reshape(-1, 1)))  # Add labels as last column
miu = 0.04
w = l_algo(X_with_labels, miu)
print("Learned weights:", w)
plot_decision_boundary(X, y, w)
```
